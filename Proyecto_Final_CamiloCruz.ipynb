```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Renombrar las columnas para mayor claridad
df.rename(columns={'Ventas $ (EPOS) (Anterior)': 'Ventas_2021', 'Ventas $ (EPOS) (Reciente)': 'Ventas_2022'}, inplace=True)

# Crear un DataFrame para las ventas en valor por región y año
df_ventas_por_region_y_anio = df.groupby(['Región'])[['Ventas_2021', 'Ventas_2022']].sum().reset_index()

# Melt para convertir las columnas de año en una única columna 'Año' y 'Ventas $ (EPOS)'
df_ventas_por_region_y_anio_melted = df_ventas_por_region_y_anio.melt(id_vars=['Región'], var_name='Año', value_name='Ventas $ (EPOS)')

# Crear la gráfica de barras
plt.figure(figsize=(12, 6))
sns.barplot(x='Región', y='Ventas $ (EPOS)', hue='Año', data=df_ventas_por_region_y_anio_melted, palette='viridis')

# Configurar la leyenda y el título
plt.legend(title='Año')
plt.title('Ventas en Valor por Región y Año')
```

```python
# Renombrar las columnas para mayor claridad
df.rename(columns={'Ventas $ (EPOS) (Anterior)': 'Ventas_2021', 'Ventas $ (EPOS) (Reciente)': 'Ventas_2022'}, inplace=True)

# Crear un DataFrame para las ventas en valor por región y año
df_ventas_por_region_y_anio = df.groupby(['Región'])[['Ventas_2021', 'Ventas_2022']].sum().reset_index()

# Calcular la variación porcentual
df_ventas_por_region_y_anio['Variacion_Porcentual'] = ((df_ventas_por_region_y_anio['Ventas_2022'] - df_ventas_por_region_y_anio['Ventas_2021']) / df_ventas_por_region_y_anio['Ventas_2021']) * 100

# Redondear la variación porcentual a dos decimales
df_ventas_por_region_y_anio['Variacion_Porcentual'] = df_ventas_por_region_y_anio['Variacion_Porcentual'].round(2)


# Crear un cuadro con la variación porcentual por región
cuadro_variacion_porcentual = df_ventas_por_region_y_anio[['Región', 'Variacion_Porcentual']].copy()

# Mostrar el cuadro
print(cuadro_variacion_porcentual)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
# Calculo del coeficinete de correlación entre ventas valor y ventas en unidades
correlation_value = df['Ventas $ (EPOS) (Reciente)'].corr(df['Unidades (EPOS) (Reciente)'])

# Gráfica de dispersión
sns.scatterplot(x='Ventas $ (EPOS) (Reciente)', y='Unidades (EPOS) (Reciente)', data=df)

slope, intercept, _, _, _ = linregress(df['Ventas $ (EPOS) (Reciente)'], df['Unidades (EPOS) (Reciente)'])

sns.lineplot(x=df['Ventas $ (EPOS) (Reciente)'], y=slope * df['Ventas $ (EPOS) (Reciente)'] + intercept, color='red',
             label=f'Correlación: {correlation_value:.2f}')

```

Existe cierta correlación entre ambas variables, sin embargo, esta no es extremadamente fuerte. Nos muestra que a medida que las ventas en unidades aumnentan, ventas en valor también lo hacen.

```python
# Dividir los datos por región
regiones_unicas = df['Región'].unique()

# Crear una figura con subgráficos para cada región
fig, axs = plt.subplots(nrows=len(regiones_unicas), figsize=(8, 4 * len(regiones_unicas)), sharex=True)

# Iterar sobre cada región y graficar la relación y la línea de regresión
for i, region in enumerate(regiones_unicas):
    subset = df[df['Región'] == region]

     # Calcular el coeficiente de correlación
    correlation_value = subset['Ventas $ (EPOS) (Reciente)'].corr(subset['Unidades (EPOS) (Reciente)'])

     # Crear una gráfica de dispersión para la región actual
    sns.scatterplot(x='Ventas $ (EPOS) (Reciente)', y='Unidades (EPOS) (Reciente)', data=subset, ax=axs[i])

    # Obtener la pendiente e intersección de la línea de regresión
    slope, intercept, _, _, _ = linregress(subset['Ventas $ (EPOS) (Reciente)'], subset['Unidades (EPOS) (Reciente)'])

     # Agregar la línea de regresión y la ecuación de la correlación
    sns.lineplot(x=subset['Ventas $ (EPOS) (Reciente)'], y=slope * subset['Ventas $ (EPOS) (Reciente)'] + intercept, color='red',
                 label=f'Correlación: {correlation_value:.2f}', ax=axs[i])

    # Configurar el título y la leyenda
    axs[i].set_title(f'Región: {region}')
    axs[i].legend()

  # Ajustar el diseño y mostrar la gráfica
plt.tight_layout()
```
