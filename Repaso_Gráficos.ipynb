```python
# cuando yo quiera comenzar a dibujar sobre estos lienzos lo que debo hacer es empezar capa por capa
# en este ejemplo de R voy a comenzar a hacer el dibujo sobre la base de datos de asesinatos o murders
# existen 2 maneras de llamar los gráficos con R
# opción 1 - %%R - ggplot(data = murders)
# opcion 2 - %%R - murders %>% ggplot()

#ambas están bien, pero si las ejecuto así tal cual no me va a dibujar nada
```

```python
# para que el dibujo tome sentido tengo que traer las capas
# entonces pegado a lo que ya tenía agrego el signo + para traer una geometría (que tipo de gráfico quiero imprimi)
# tipos de geometría son puntos, lineas, barras
%%R
ggplot(data = murders) +
  geom_point(mapping = aes (x = population,y = total))
```

```python
# todo gráfico debe tener sí o sí el elemento estético, porque de ahí es que le da valor al eje x y al eje y
# mapping
# aes = estética
%%R
ggplot(data = murders) +
  geom_line(mapping = aes (x = population,y = total))

  # cuando hago una geometría de puntos o lineas uso dos variables estéticas x y y
  # cuando es de barras geom_bar la estética solo tendrá una variable x
```

```python
 # cuando hago una geometría de puntos o lineas uso dos variables estéticas x y y
 # cuando es de barras geom_bar la estética solo tendrá una variable x

%%R

ggplot(data = murders) +
  geom_bar(mapping = aes (x = region))
```

```python
# en las gráfias se pueden mezclar las geometrías para tener gráficos más complejos
```

```python
# acá ya estoy mezclando dos geometrías en un solo gráfico
# es importante decir que puedo aplicar estéticas como filtros y como subfiltros
%%R

ggplot(gapminder[gapminder$country == "Albania",], aes (x=year, y =life_expectancy)) +
      geom_line(aes(group=1))+
       geom_point()
```

```python
# el gráfico de bigotes es el más importante que debo aprender a usar por toda la información que me da

%%R

ggplot(murders, aes(x= region, y=total)) +
  geom_boxplot()
```

```python
# las geometrías en python son de dos tipos pnt y sns
# la diferencia es que uno genera gráficas planas y el otro las hace dinámicas o interactivas
```

```python
# los gráficos de barras en python trabajan un concepto que se llama bins y es básicamente la distancia que
# hay entre el observador y el gráfico entre mayor sea el número desde más lejos se está viendo

plt.hist(df1["Género"], bins=6)

```

```python
# ahora si quiero ordenarlo mejor se hace de la siguiente manera
# cuando me refiero a ordenarlo mejor, es ponerle nombre a los ejes

plt.hist(df1["Edad del Padre"], bins=6)
plt.xlabel("Edad del padre")
plt.ylabel("no sé")


# en python los labels no tienen que ser variables del DF, son los nombres que yo deseo colocarle a los ejes
# ojo los histogramas son para varialbles cuantitativas (números)
```
